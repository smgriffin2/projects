```python
# make lists
demo_dict = {
    "pa": ["democrat", "republican", "politcally-moderate"],
    "race": ["black", "white", "asian", "hispanic", "other"],
    "sex": ["male", "female"]
}
# make all combos of list items as list
# insert all combos into prompt
demo_dict
```

```python
# Make all combo of dict of lists:
demo_inputs = list(itertools.product(*demo_dict.values()))
```

```python
len(demo_inputs[0])
```

```python
prompts = []
def prompt_str(list=demo_inputs):
    for item in list:
        if "other" in item:
            continue
        prompts.append("write a 75 word message for a {}, {}, {} urging them to support the green new deal"\
              .format(item[0], item[1], item[2]))
```

```python
prompt_str()
```

```python
prompts
```

```python
responses = []
for item in prompts:
    print(item)
    response = openai.Completion.create(
      model="text-davinci-003",
      prompt=item,
      max_tokens=100,
      temperature=1,
        n=3
        
)
    responses.append(response)
responses
```

```python
for item in responses:
    for i in item["choices"]:
        print(i["text"])
```

## MINI TEST

```python
dem_responses = []
rep_responses = []
mod_responses = []
for item in prompts:
    print(item)
    response =  openai.Completion.create(
      model="text-davinci-003",
      prompt=item,
      max_tokens=100,
      temperature=1.5,
        n=3)
    if "democrat" in item:
        dem_responses.append(response)
    elif "republican" in item:
        rep_responses.append(response)
    else:
        mod_responses.append(response)
```

```python
dem_response_list = []
for item in dem_responses:
    for i in item["choices"]:
        dem_response_list.append(i["text"])
        
dem_response_list
```
