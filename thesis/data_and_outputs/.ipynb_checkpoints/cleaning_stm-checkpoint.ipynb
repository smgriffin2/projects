```python
# Create dict with all prompts
prompts = {"dem-tree":'Persuade Democrat support planting trees offset carbon emissions.',
 "dem-tax":'Persuade Democrat support giving tax credits businesses engaging carbon capture storage.',
 "dem-fuel":'Persuade Democrat support tougher fuel efficiency standards cars.',
 "rep-tree":'Persuade Republican support planting trees offset carbon emissions.',
 "rep-tax":'Persuade Republican support giving tax credits businesses engaging carbon capture storage.',
 "rep-fuel":'Persuade Republican support tougher fuel efficiency standards cars.'}

# process prompts
for prompt in prompts:

    prompts[prompt] = prompts[prompt].lower()
    prompts[prompt] = prompts[prompt].replace(".", "")
    prompts[prompt] = prompts[prompt].split(" ")

for i in range(len(df_para)):
    query = df_para.loc[i, "text"]
    for key in prompts:
        if df_para.loc[i, "prompt"] == key:
            stopwords = prompts[key]
            querywords = query.split()
            resultwords  = [word for word in querywords if word.lower() not in stopwords]
            result = ' '.join(resultwords)
            df_para.loc[i, "processed_text"] = result
```
